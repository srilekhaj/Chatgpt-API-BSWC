```python
import os
from dotenv import load_dotenv
load_dotenv()
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

response = client.responses.create(
    model="gpt-5-nano",
    input=[
        {"role": "system", "content": "You are a healer mind ."},
        {"role": "user", "content": "Write a one-sentence to me personally touching my heart."}
    ]
)

print(response)

```

```python
print(response.output_text)
print(response.usage.input_tokens)
print(response.usage.output_tokens)
print(response.usage.total_tokens)
```

```python
[
  {"role": "system", "content": "You are a helpful coding assistant."},
  {"role": "user", "content": "Write a Python function to reverse a string."},
  {"role": "assistant", "content": "Sure! Here's a function:\n\ndef reverse_string(s):\n    return s[::-1]"}
]
The system sets the context. behaviour of assistant.

The user asks something.

The assistant replies.
```

The user sends a request to the OpenAI API.

The system role sets instructions, rules, or persona for the assistant.

The assistant (AI model) processes the request within that context.

The API returns the response back to the user.

```python
   ┌─────────────┐
   │   User      │
   │ (human input)│
   └─────┬───────┘
         │ sends request
         ▼
   ┌─────────────┐
   │   API       │
   │ (OpenAI)    │
   └─────┬───────┘
         │
         │ interprets roles
         ▼
   ┌─────────────┐
   │ System Role │
   │ Sets context│
   └─────┬───────┘
         │ guides behavior
         ▼
   ┌─────────────┐
   │ Assistant   │
   │ (AI model)  │
   └─────┬───────┘
         │ generates response
         ▼
   ┌─────────────┐
   │   User      │
   │ Receives AI │
   │ Response    │
   └─────────────┘
```
